```python
# 오후수업 
# 임의 x,y값 제작 

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(100)

# 표준정규분포의 임의데이터셋 제작 
x=np.random.randn(1000)
y=np.random.randn(1000)

plt.scatter(x,y)
# plt.hist(x),np.std(x)
```

```python
# 
w=0.2 ; b=0.1
y_hat=w*x+b

plt.scatter(x,y)
plt.plot(x,y_hat)

# 평균제곱오차(MSE) 
MSE=np.mean((y-y_hat)**2)
plt.title(MSE)
```

```python
from os import error


plt.figure(figsize=(20,6))
wData=[0.2,0.1,0.3,2]
bData=[0.1,0.1,0,-0.5]

cnt=0
for w,b in zip(wData,bData):
    cnt+=1
    plt.subplot(1,5,cnt)
    
    y_hat=w*x+b

    plt.ylim([-8,8])
    plt.scatter(x,y)
    plt.plot(x,y_hat)

    MSE=np.mean((y-y_hat)**2)
    plt.title(MSE)

plt.show()

```

```python
### w,b값을 계속 갱신하면서 mse를 확인해보겠음
### 임의로 w,b값을 갱신해보겠음
x=np.random.randn(1000)
y=np.random.randn(1000)

plt.figure(figsize=(20,6))
cnt=0
w=2 ; b=0.1
lr=0.0001

epochs=10

for xx in range(epochs): 
    cnt+=1
    plt.subplot(1,epochs,cnt)

    y_hat=w*x+b
    
    plt.ylim([-8,8])
    plt.scatter(x,y)
    plt.plot(x,y_hat)

    err=np.mean((y-y_hat)**2)
    plt.title(err)
    
    w=w-w*err*lr
    b=b-b*err*lr
plt.show()
```
